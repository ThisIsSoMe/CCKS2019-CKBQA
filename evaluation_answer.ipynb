```python
# 单个问句的prf值
def one_value(pred,gold):
    pred=set(pred)
    gold=set(gold)
    inter=pred.intersection(gold)
    if len(inter)==0:
        p,r,f=0.0,0.0,0.0
    else:
        p=float(len(inter)/len(pred))
        r=len(inter)/len(gold)
        f=2*p*r/(p+r)
    return p,r,f
```

```python
fn_ans = ['answer_file_name.json']
all_ans={}
for fn in fn_ans:
    with open(fn,'r')as f:
        one_dict=json.load(f)
        for k,v in one_dict.items():
                all_ans[k]=v
```
