## 训练集数据生成

```python
neg_size = 10
max_neg = 200
cur_data = train_all_data
count = 0
fn_out_train = 'train.json' 
with open(fn_out_train, 'w')as fout:
    
    all_questions = []
    all_golds = []
    all_negs = []
    
    for que,label,ent,tri,mentions,ents in zip(cur_data['questions'],cur_data['types'],cur_data['ents'],cur_data['triples'],cur_data['mentions'],cur_data['ents']):
        if label == 1:
            cur_ent = tri[0]
            cur_ment = del_des(cur_ent)
            gold_tri = (cur_ent if des else cur_ment, tri[1], PAD)
        elif label == 2:
            cur_ent = tri[1]
            cur_ment = del_des(cur_ent)
            gold_tri = (PAD, tri[0], cur_ent if des else cur_ment)
        elif label == 4 or label == 5:
            cur_ent = tri[0]
            cur_ment = del_des(cur_ent)
            gold_tri = (cur_ent if des else cur_ment, tri[1], tri[2], PAD)
        elif label ==7:
            cur_ent = tri[1]
            cur_ment = del_des(cur_ent)
            gold_tri = (cur_ent if des else cur_ment, tri[0], tri[2], PAD)
        else:
            continue
        
        mentions = [e[1:-1] for e in ents]
        
        
        neg_paths1 = get_one_hop_path(cur=cur, mentions=mentions, PAD='<E>', topk=max_neg, addition=cand_relations, addition_bound=neg_size,gold=gold_tri, des=des)

        neg_paths2 = get_two_hop_path(cur=cur, mentions=mentions, PAD='<E>', topk=max_neg, addition=cand_relations, addition_bound=neg_size,gold=gold_tri, des=des)
        
#             from pdb import set_trace
#             set_trace()

        neg_paths = neg_paths1 + neg_paths2
        
        if neg_paths:
            count += 1
            all_questions.append(que)
            all_golds.append(''.join(list(gold_tri)))
            all_negs.append([''.join(path) for path in neg_paths])
            print("num %s q, %d neg paths"%(count, len(neg_paths)))
            
    
    # dump
    all_train_data = {}
    all_train_data['questions'] = all_questions
    all_train_data['golds'] = all_golds
    all_train_data['negs'] = all_negs
    
    json.dump(all_train_data, fout, ensure_ascii=False)
```
